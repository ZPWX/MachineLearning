```python
from numpy import *
from pylab import *
import matplotlib.pyplot as plt
import numpy as np
#
A = np.mat("0 1 2;1 0 3;4 -3 8")
print("A\n",A)
inverse = np.linalg.inv(A)
print("Inverse of A\n",inverse)
print("Check\n", A*inverse)
#
B = np.mat("1 -2 1;0 2 -8;-4 5 9")
print("B\n",B)
b = array([0,8,-9])
print("b\n",b)
x = np.linalg.solve(B,b)
print("Solution",x)
print("Check\n",np.dot(B,x))
#
C = np.mat("3 -2;1 0")
print("C\n",C)
#
print("Eigenvalues",np.linalg.eigvals(C))
eigenvalues,eigenvectors = np.linalg.eig(C)
print("First tuple of eig\n",eigenvectors)
print("Second tuple of eig\n",eigenvectors)
#dotAx=ax
for i in range(len(eigenvectors)):
    print("Left\n",np.dot(C,eigenvectors[:,i]))
    print("Right\n",eigenvectors[i]*eigenvectors[:,i])
    print("Done")
#
D =  np.mat("4 11 14;8 7 -2")
print("D\n",D)
U,sigma,V = np.linalg.svd(D,full_matrices=False)
print("U\n",U)
print("sigma\n",sigma)
print("V\n",V)
#
pseudoinv = np.linalg.pinv(D)
print("Pseudo inverse\n",pseudoinv)
#
E = np.mat("3 4;5 6")
print("E\n",E)
print("Determainant\n",np.linalg.det(E))
#
x1 = np.linspace(0, 2*np.pi, 30) #30
wave = np.cos(x1)
transformed = np.fft.fft(wave)  #fft
shifted = np.fft.fftshift(transformed)
print(np.all(np.abs(np.fft.ifft(transformed) - wave) < 10 ** -9)) #ifft
print(np.all((np.fft.ifftshift(shifted) - transformed) < 10 ** -9)) #ifftshift
print("Done")
plot(transformed)  #Matplotlib
show()
plot(transformed, lw=2)  #
plot(shifted, lw=3)      #
show()
print("Done")
#
cash = np.zeros(10000)
cash[0] = 1000
outcome = np.random.binomial(9,0.5,size=len(cash))

for i in range(1,len(cash)):
    if outcome[i] < 5:
        cash[i] = cash[i - 1] - 1
    elif outcome[i] < 10:
        cash[i] = cash[i - 1] + 1
    else:
        raise AssertionError("Unexpected outcome" + outcome)
print(outcome.min(), outcome.max())
plot(np.arange(len(cash)), cash)
show()
print("Done")
#
''''''
points = np.zeros(100)
outcomes = np.random.hypergeometric(25,1,3,size=len(points))

for i in range(len(points)):
    if outcomes[i] == 3:
        points[i] = points[i - 1] + 1
    elif outcomes[i]  == 2:
        points[i] = points[i - 1] - 6
    else:
        print(outcomes[i])
plot(np.arange(len(points)),points)
show()
print("Done")
#
'''
N = 1000
normal_values = np.random.normal(size=N) #normal_values1000
dummy,bins,dummy = plt.hist(normal_values, np.sqrt(N),normed=True,lw=1)
sigma = 1
mu = 0
plt.plot(bins, 1/(sigma*np.sqrt(2*no.pi))*np.exp(-(bins - mu)**2/(2*sigma**2)),lw=2)
plt.show()

#
N = 1000
lognormal_values = np.random.lognormal(size=N)
dummy, bins, dummy = plt.hist(lognormal_values, np.sqrt(N), normed=True, lw=1)
sigma = 1
mu =0
x = np.linspace(min(bins), max(bins), len(bins))
pdf = np.exp(-(numpy.log(x) - mu)**2/(2*sigma**2))/(x*sigma*np.sqrt(2*np.pi))
plt.plot(x, pdf, lw=3)
plt.show()
'''
#
np.random.seed(42)
complex_numbers = np.random.random(5) + 1j*np.random.random(5)
print("Complex numbers\n",complex_numbers)
print("Sorted\n",np.sort_complex(complex_numbers)) #sort_complex
#searchsorted
''''''
a = np.arange(5)
indices = np.searchsorted(a,[-2,7])
print(a)
print("Indices\n",indices)
print("The full array\n",np.insert(a, indices, [-2, 7]))
#
condition = (a % 2) == 0 #
print("Even numbers\n",np.extract(condition, a)) #
print("Non zero\n",np.nonzero(a))  #nonzero
#
window = np.bartlett(42)
plot(window)
show()
```
